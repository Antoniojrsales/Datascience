```python
#Biblioteca
import pandas as pd
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = load_iris() #Carregando o conjunto de dados Iris
x_data = pd.DataFrame(iris['data'], columns=iris['feature_names']) #Transformando o conjunto em DataFrame Pandas
x_data['target'] = iris['target']
```

```python
sns.pairplot(x_data, hue='target', diag_kind='hist')
plt.show()
```

```python
for col in iris['feature_names']:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='target', y=col, data=x_data)
    plt.title(f'Boxplot of {col} by target class')
    plt.show()
```

```python
correlacao_target = x_data.corr()
```

```python
#-Gere um heatmap para identificar quais variveis esto mais correlacionadas.
plt.figure(figsize=(12, 8))
plot = sns.heatmap(
    correlacao_target, 
    annot=True,
    cmap='coolwarm', 
    linewidths=0.5,
)
plt.title('Heatmap das Correlaes - Iris Setosa', fontsize=14)
plt.show()
```

```python
correlacao_target.plot(kind='bar')
plt.title('Correlao das Variveis com Target')
plt.xlabel('Variveis')
plt.ylabel('Correlao')
plt.show()
```

```python
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='target', data=x_data)
plt.title('Sepal Length vs Sepal Width')
plt.show()
```
